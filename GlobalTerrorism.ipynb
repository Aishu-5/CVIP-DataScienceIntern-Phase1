**Number of Attacks each year**

```python
x_year = terrorism['Year'].unique()
y_count_years = terrorism['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

**Total Number of people killed in terrorist attack**

```python
killedData = terrorism.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killedData.dropna())))
```

```python
attackedData = terrorism.loc[:,'AttackType']
typeKilledData = pd.concat([attackedData, killedData], axis=1)
```

```python
typeKilledData.head()
```

```python
typeKilledFormatData = typeKilledData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKilledFormatData
```

```python
typeKilledFormatData.info()
```

**Number of Killed in Terrorist Attacks by Countries**

```python

countryData = terrorism.loc[:,'Country']
# countyData
countryKilledData = pd.concat([countryData, killedData], axis=1)
```

```python
countryKilledFormatData = countryKilledData.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKilledFormatData
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
```

**Number of people killed by countries**

```python
labels = countryKilledFormatData.columns.tolist()
labels = labels[:50] #50 bar provides nice view
index = np.arange(len(labels))
transpoze = countryKilledFormatData.T
values = transpoze.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values] # convert float to int
colors = ['red', 'green', 'blue', 'purple', 'yellow', 'brown', 'black', 'gray', 'magenta', 'orange'] # color list for bar chart bar color
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=20)
plt.xlabel('Countries', fontsize = 20)
plt.xticks(index, labels, fontsize=18, rotation=90)
plt.title('Number of people killed by countries', fontsize = 20)
# print(fig_size)
plt.show()
```
