**Separating X and Y data**

```python
x=df.drop(['diagnosis'],axis=1)
y=df.diagnosis
```

**Splitting the data into training and testing data**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

**Standardizing the training and testing data**

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler()
x_train=standard_scaler.fit_transform(x_train)
x_test=standard_scaler.transform(x_test)
```

**Classification Models**

```python
key = ['DecisionTreeClassifier','LogisticRegression','KNeighborsClassifier','RandomForestClassifier','SVC','AdaBoostClassifier']
value = [ DecisionTreeClassifier(criterion="entropy",random_state=42),LogisticRegression(), KNeighborsClassifier(n_neighbors = 9, weights ='uniform'),RandomForestClassifier(n_estimators=300), SVC(kernel="rbf",random_state=15), AdaBoostClassifier(n_estimators=100)]
models = dict(zip(key,value))
models
```

**Accuracies of the classification models**

```python
predicted =[]
for name,algo in models.items():
    model=algo
    model.fit(x_train,y_train)
    predict = model.predict(x_test)
    acc = accuracy_score(y_test, predict)*100
    predicted.append(acc)
    print(name,acc)
```

**Plotting the Model Accuracies**

```python
plt.figure(figsize = (10,5))
sns.barplot(x = predicted, y = key, palette='PuRd')
plt.title("Plotting the Model Accuracies", fontsize=16, fontweight="bold")
```
